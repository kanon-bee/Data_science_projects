So we will change all columns name from camelcase to snakecase first and then shorten some of the columns name 


```python
a.columns
```

```python
new_columns = []
```

```python
def change_case(str): 
    res = [str[0].lower()] 
    for c in str[1:]: 
        if c in ('ABCDEFGHIJKLMNOPQRSTUVWXYZ'): 
            res.append('_') 
            res.append(c.lower()) 
        else: 
            res.append(c) 
      
    return ''.join(res)
```

```python
for i in a.columns:
    b = change_case(i)
    new_columns.append(b)
```

```python
a.columns = new_columns
```

# Finding the most common make/model combinations

Next, We will figure out the most common make/model comnbinations

```python
top_brand = a['brand'].value_counts().head(5)
```

```python
top_brand
```

```python
model_freq_vw = {}
```

```python
import operator
vw = a['brand'] == 'volkswagen'
vw_models = a[vw]['model']

model_freq_vw = {}

for m in vw_models:
    if m not in model_freq_vw:
        model_freq_vw[m] = 1
    else:
        model_freq_vw[m] += 1
sorted_vw = sorted(model_freq_vw.items(), key=operator.itemgetter(1), reverse=True)
sorted_vw[0]
```

```python
def brand_model_freq(b):
    
    brand_filter = a['brand'] == b
    brand_models = a[brand_filter]['model']
    
    model_freq = {}
    
    for m in brand_models:
        if m not in model_freq:
            model_freq[m] = 1
        else:
            model_freq[m] += 1
    
    sorted_brand = sorted(model_freq.items(), key=operator.itemgetter(1), reverse=True)
    
    return sorted_brand[0]
```

```python
audi = brand_model_freq('audi')
```

```python
opel = brand_model_freq('opel')
```

```python
bmw = brand_model_freq('bmw')
```

```python
mercedes_benz = brand_model_freq('mercedes_benz')
```

```python
volkswagen = brand_model_freq('volkswagen')
```
